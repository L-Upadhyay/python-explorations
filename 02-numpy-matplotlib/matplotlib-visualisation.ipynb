# Matplotlib Visualisation Techniques

Five plots covering core Matplotlib patterns:

1. **E = mc²** — translating a physics equation into a line plot
2. **E = mc² (log scale)** — logarithmic y-axis with axis-selective grid
3. **US Treasury yield curves** — dual y-axis plot comparing 2007 vs 2020
4. **subplot2grid layout** — asymmetric subplot arrangement (sine, exp decay, log curve)

All plots use the `fig, ax` object-oriented API — more flexible and reusable than `plt.*` function calls.

```python
import numpy as np
import matplotlib.pyplot as plt
```

---
## 1. E = mc² — Linear Plot

Translating Einstein's equation directly into arrays. `m` is 11 evenly spaced points from 0–10g. `E = m × c²` applies element-wise — no loop needed because NumPy broadcasts the scalar `c²` across the entire array.

```python
c = 3e8                        # speed of light (m/s)
m = np.linspace(0, 10, 11)    # mass: 0 to 10 grams
E = m * (c ** 2)               # E = mc² — element-wise

print("m (grams):", m)
print("E (joules):", E)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(m, E, color='red', linewidth=2)
ax.set_title(r'$E = mc^2$', fontsize=14)
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')
ax.set_xlim(0, 10)
ax.set_ylim(0, 9e17)
ax.grid(True)

plt.tight_layout()
plt.show()
```

---
## 2. E = mc² — Logarithmic Scale

A log y-axis is useful when values span multiple orders of magnitude. Here it reveals the curve's shape more clearly at small masses.

Two key additions from the linear version:
- `ax.set_yscale('log')` — switches the y-axis to base-10 logarithmic spacing
- `ax.grid(True, axis='y')` — grid lines only on y-axis ticks (cleaner than full grid)

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(m, E, color='red', linewidth=2)
ax.set_title(r'$E = mc^2$ (Logarithmic Scale)', fontsize=14)
ax.set_xlabel('Mass (grams)')
ax.set_ylabel('Energy (joules)')

ax.set_yscale('log')           # logarithmic y-axis
ax.grid(True, axis='y')        # grid on y-axis ticks only
ax.set_xlim(0, 10)
ax.set_ylim(1e17, 1e18)

plt.tight_layout()
plt.show()
```

## 3. US Treasury Yield Curves — Dual Y-Axis

Plotting the July 16th yield curve for 2007 (pre-financial crisis) and 2020 (COVID-19 era) on the same chart. The two datasets have very different scales (2007 rates ~5%, 2020 rates ~0.1–1.3%), so a **dual y-axis** (`twinx`) is used — each curve gets its own y-scale and y-label, allowing both to be visible without distortion.

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

x        = np.arange(len(labels))
y_2007   = np.array(july16_2007)
y_2020   = np.array(july16_2020)

print(f"2007 range: {y_2007.min():.2f}% – {y_2007.max():.2f}%")
print(f"2020 range: {y_2020.min():.2f}% – {y_2020.max():.2f}%")
print("Scale ratio:", round(y_2007.mean() / y_2020.mean(), 1), "× — justifies dual y-axis")
```

```python
fig, ax1 = plt.subplots(figsize=(11, 5))

# 2007 — left y-axis (blue)
color_2007 = 'blue'
ax1.plot(x, y_2007, color=color_2007, linewidth=2)
ax1.set_xlabel('Contract Length')
ax1.set_ylabel('2007', color=color_2007, fontsize=12)
ax1.tick_params(axis='y', labelcolor=color_2007)
ax1.set_xticks(x)
ax1.set_xticklabels(labels)

# 2020 — right y-axis (red)
ax2 = ax1.twinx()
color_2020 = 'red'
ax2.plot(x, y_2020, color=color_2020, linewidth=2)
ax2.set_ylabel('2020', color=color_2020, fontsize=12)
ax2.tick_params(axis='y', labelcolor=color_2020)

plt.title('July 16th Yield Curves', fontsize=14)
plt.tight_layout()
plt.show()

print("""
Insight: The 2007 yield curve is nearly flat (inverted at short durations —
a classic recession signal). The 2020 curve is steep and much lower overall,
reflecting the Fed's near-zero interest rate response to COVID-19.
""")
```

## 4. Asymmetric Subplot Layout with `subplot2grid`

`subplot2grid` allows non-uniform subplot sizes within a single figure — unlike `plt.subplots()` which only creates equal-sized grids.

Layout: 2×2 grid where the right subplot spans both rows (`rowspan=2`), giving three plots of two different sizes.

```python
x      = np.linspace(0, 10, 100)
y_sin  = np.sin(x)
y_exp  = np.exp(-x)     # e^(-x): decays from 1 toward 0
y_log  = np.log(x + 1)  # log(x+1): avoids log(0) at x=0
```

```python
fig = plt.figure(figsize=(10, 8))

# Top-left: Sine Wave — position (row=0, col=0) in a 2×2 grid
ax1 = plt.subplot2grid((2, 2), (0, 0))
ax1.plot(x, y_sin, color='blue')
ax1.set_title('Sine Wave')
ax1.grid(True)

# Bottom-left: Exponential Decay — position (row=1, col=0)
ax2 = plt.subplot2grid((2, 2), (1, 0))
ax2.plot(x, y_exp, color='green')
ax2.set_title('Exponential Decay')
ax2.grid(True)

# Right tall: Logarithmic Curve — starts at (row=0, col=1), spans 2 rows
ax3 = plt.subplot2grid((2, 2), (0, 1), rowspan=2)
ax3.plot(x, y_log, color='red')
ax3.set_title('Logarithmic Curve')
ax3.grid(True)

plt.suptitle('Subplots Using subplot2grid', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()

print("""
subplot2grid key arguments:
  shape    = (2, 2)  — the overall grid dimensions
  loc      = (r, c)  — top-left corner of this subplot
  rowspan  = 2       — how many rows this subplot occupies
  colspan  = 1       — how many columns (default)
""")
```
